```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
carteira = pd.read_excel('carteira2.xlsx')
display(carteira)
```

```python
quantidade = carteira['Qtde']
ativos = carteira['Ativos']

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(quantidade, labels=ativos, autopct='%1.1f%%', startangle=140)
plt.title('Porcentagem de cada ativo na carteira')
plt.show()
```
